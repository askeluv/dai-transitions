```python
import pandas as pd
from lifelines import KaplanMeierFitter
import matplotlib
import sys
sys.path.append("..") # so we can import our custom utils module
import utils
from functools import reduce
%matplotlib inline
matplotlib.rcParams['figure.dpi'] = 100
```

```python
df = pd.read_excel('../data/debt_lives_processed.xlsx')
```

```python
sequences = utils.dataframe_to_sequences(df)
sequences[:5] # see first 5 CDPs represented as state sequences
```

```python
num_states = pd.Series(map(len, sequences))
last_state = pd.Series(map(lambda x: x[-1][0], sequences))
age = pd.Series(map(lambda x: x[-1][1] - x[0][1] if len(x) > 1 else utils.LAST_TIMESTAMP - x[0][1], sequences))
age = age / (60 * 60 * 24) # convert to hours
is_censored = pd.Series(map(lambda x: len(x) == 1, sequences))
is_dead = pd.Series(map(lambda x: int(x[-1][0] in ('wiped', 'shut', 'bitten')), sequences))
born_month = pd.Series(map(lambda x: pd.Timestamp.fromtimestamp(x[0][1]).month, sequences))
```

```python
cdps = pd.concat([num_states, last_state, age, is_censored, is_dead, born_month], axis=1)
cdps.columns = ['num_states', 'last_state', 'age', 'is_censored', 'is_dead', 'born_month']
cdps.head(5)
```

```python
T, E = cdps['age'], ~cdps['is_censored']
model = KaplanMeierFitter().fit(T, E)
ax = model.plot(figsize=(6, 4))
ax.set_xlabel('Days');
ax.set_ylabel('Survival Rate');
ax.set_title('CDP Survival Rate');
```

```python
ax = matplotlib.pyplot.subplot(111)
ax.set_title('CDP Survival Curves: Monthly Cohorts')
for month in cdps['born_month'].unique():
    loc = cdps[cdps['born_month'] == month].index
    tmp_cdps = cdps.loc[loc]
    T, E = tmp_cdps['age'], ~tmp_cdps['is_censored']
    model = KaplanMeierFitter().fit(T, E)
    model.plot(ax=ax, figsize=(6, 4), label=month);
    ax.set_xlabel('Days');
    ax.set_ylabel('Survival Rate');
```
